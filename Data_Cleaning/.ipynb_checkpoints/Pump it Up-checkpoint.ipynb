## 根据0值和缺失值比例排列，从低到高开始处理缺失值

#### 1. longitude（经度）            0.030505
这显然是一个位置的特征，本数据集中与位置相关的特征，最精确的应该是’subvillage‘，先尝试下subvillage，后发现是同一种subvillage  
发生的缺失longitude，所以上移一层到ward,还是存在问题，再上移一层到lga

```python
all_df_dropped[all_df_dropped['longitude'].isna()]['region']
# 全都没有缺失
# all_df_dropped[all_df_dropped['region'] == 'Shinyanga']
filtered_df = all_df_dropped[(all_df_dropped['longitude'] != 0) & (all_df_dropped['longitude'].notna())]
# filtered_df[all_df_dropped['region'] == 'Shinyanga']
grouped_df = filtered_df.groupby('lga')['longitude'].mean()
grouped_df
```

```python
# long_ins = []
# for lga, mean_long in grouped_df.items():
#     # 选择对应的 lga 且 longitude 为 0 的行
#     condition = (all_df_dropped['lga'] == lga) & (all_df_dropped['longitude'] == 0)
#     # 替换这些行的 longitude 值为该 lga 的平均 longitude 值
#     all_df_dropped.loc[condition, 'longitude'] = mean_long
#     long_ins.append(mean_long)

# # 保存修改后的 DataFrame 为 all_df1
# all_df1 = all_df_dropped
```

```python
# 在 all_df_dropped 中添加一个新列以追踪哪些 longitude 被更新
all_df_dropped['longitude_updated'] = False  # 默认所有行都未被更新

# 遍历 grouped_df 来替换并标记被更新的 longitude
for lga, mean_long in grouped_df.items():
    # 选择对应的 lga 且 longitude 为 0 的行
    condition = (all_df_dropped['lga'] == lga) & (all_df_dropped['longitude'] == 0)
    # 替换这些行的 longitude 值为该 lga 的平均 longitude 值，并标记为已更新
    all_df_dropped.loc[condition, 'longitude'] = mean_long
    all_df_dropped.loc[condition, 'longitude_updated'] = True  # 标记为已更新

# 查看所有被更新的行
updated_rows = all_df_dropped[all_df_dropped['longitude_updated'] == True]

# 保存修改后的 DataFrame 为 all_df1
all_df1 = all_df_dropped

# 查看被更新的行
print(updated_rows[['lga', 'longitude', 'longitude_updated']])
```

```python
# 缺失值的比例
# all_df.isna().sum()
all_df1.isna().sum()

missing_values_count = all_df1.isna().sum()

total_rows = all_df1.shape[0]

missing_values_ratio = missing_values_count / total_rows

missing_values_ratio.sort_values(ascending=True, inplace=True)

missing_values_ratio
```
